```python
X = []
Y = []
x,y = [], []
names = []
for n in range (len(wav)):
  names.append(str(n))
print(names)
```

```python
def record_to_mfcc():
  print('start')
  samplerate, data = wavfile.read(wav[i])
  x = data.astype('float32')
  # x = x[0:8000]
  return librosa.feature.mfcc(x)
```

```python
for i in range(len(wav)):
  for o in range(5):
    a = record_to_mfcc()
    print(a.shape)
    x.append(a)
    y.append(names.index(names[i]))
  X.append(np.array(x))
  Y.append(np.array(y))
  x, y = [], []
```

```python
model = keras.Sequential()
for a in range (44):
  print(a)
  model.add(keras.layers.Conv2D(64, 5, input_shape=(20, X[a].shape[2], 1), activation='relu'))
  model.add(keras.layers.Dropout(0.5))
  model.add(keras.layers.MaxPool2D())
  model.add(keras.layers.Conv2D(64, 5, activation='relu'))
  model.add(keras.layers.Dropout(0.5))
  model.add(keras.layers.MaxPool2D())
  model.add(keras.layers.Flatten())
  model.add(keras.layers.Dense(1024, activation='relu'))
  model.add(keras.layers.Dropout(0.5))
  model.add(keras.layers.Dense(1, activation='softmax'))
  model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
  model.fit(X[a][:,:,:,None], Y[a], epochs=100)
```

```python
import random

for q in range(30):
  i = random.randint(0,43)
  print(i)
  x = record_to_mfcc()
  print(names[model.predict(x[None,:,:,None]).argmax()])
```
